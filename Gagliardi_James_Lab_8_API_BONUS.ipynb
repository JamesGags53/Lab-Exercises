```python
import requests
#In this example, I use the NHL's official website to gather data and use various API calls to output various information
#Example API call to get information about your team of choice, in my case, my favorite team is the New York rangers, who are assigned team_id 3
team_id = "3"
team_url = f"https://statsapi.web.nhl.com/api/v1/teams/{team_id}"
team_response = requests.get(team_url)

if team_response.status_code == 200:
    team_data = team_response.json()
    print("Team data:")
    print(team_data)
else:
    print("Error:", team_response.status_code)

#Example API call to get information about an NHL game, in this case I picked a game that was in progress at the time of my coding this, being the New Jersey Devils vs the Buffalo Sabres.
game_id = "2022021148"
game_url = f"https://statsapi.web.nhl.com/api/v1/game/{game_id}/feed/live"
game_response = requests.get(game_url)

if game_response.status_code == 200:
    game_data = game_response.json()
    print("Game data:")
    print(game_data)
else:
    print("Error:", game_response.status_code)

#Example API call to get information about a specific NHL player. In this case, I found the player_id for one of my favorite players on the New York Rangers, Artemi Panarin
player_id = "8478550"
player_url = f"https://statsapi.web.nhl.com/api/v1/people/{player_id}"
player_response = requests.get(player_url)

if player_response.status_code == 200:
    player_data = player_response.json()
    print("Player data:")
    print(player_data)
else:
    print("Error:", player_response.status_code)

#Example API call to get NHL standings for a specific season, in this case in this case I picked last year's season, 2021-2022
season = "20212022"
standings_url = f"https://statsapi.web.nhl.com/api/v1/standings?season={season}"
standings_response = requests.get(standings_url)

if standings_response.status_code == 200:
    standings_data = standings_response.json()
    print("Standings data:")
    print(standings_data)
#Example API call to get NHL player stats for a specific season, in this case I use Artemi Panarin and his player_id again
season = "20202021"
player_id = "8478550"
player_stats_url = f"https://statsapi.web.nhl.com/api/v1/people/{player_id}/stats?stats=statsSingleSeason&season={season}"
player_stats_response = requests.get(player_stats_url)

if player_stats_response.status_code == 200:
    player_stats_data = player_stats_response.json()
    print(f"Player stats for Artemi Panarin for the {season} season:")
    print(player_stats_data)
else:
    print("Error:", player_stats_response.status_code)
```

```python
import requests
import time
#in this example, I create a program using the API to get the scores of current NHL games. Note that I have not tried this code while there is no active NHL games so please keep that in mind!
def get_scores():
    url = "https://statsapi.web.nhl.com/api/v1/schedule"
    response = requests.get(url)
    data = response.json()

    games = data["dates"][0]["games"]
    if len(games) == 0:
        print("No active NHL games")
    print("Current Scores:")
    for game in games:
        home_team = game["teams"]["home"]["team"]["name"]
        home_score = game["teams"]["home"]["score"]
        away_team = game["teams"]["away"]["team"]["name"]
        away_score = game["teams"]["away"]["score"]
        print(f"{away_team} {away_score} - {home_team} {home_score}")

while True:
    get_scores()
    time.sleep(30)
#the time.sleep function makes the get_scores() function reiterate every 30 seconds, updating the scores live
```
