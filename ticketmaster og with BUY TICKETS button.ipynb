```python
import requests
from tkinter import *
import tkinter as tk
from datetime import datetime
import webbrowser

tm_apikey = "Ez3L1QyVXpDzd4qTzySrYr9UcDQxuxxX"

def search_performers():
    performer_name = performer_entry.get()
    area_name = area_entry.get()
    url = 'https://app.ticketmaster.com/discovery/v2/events.json'
    params = {'keyword': performer_name, 'apikey': tm_apikey, 'city': area_name}
    response = requests.get(url, params=params)
    events = response.json()['_embedded']['events']
    sorted_events = sorted(events, key=lambda event: event['dates']['start']['localDate'])
    for event in range(0, len(sorted_events)):
        name = sorted_events[event]['name']
        if 'priceRanges' in sorted_events[event]:
            min_price = sorted_events[event]['priceRanges'][0]['min']
            max_price = sorted_events[event]['priceRanges'][0]['max']
        else:
            min_price = 'N/A'
            max_price = 'N/A'
        buy_tickets_button = Button(window, text="Buy Tickets")
        if 'url' in sorted_events[event]:
            buy_tickets_button.config(state=NORMAL, command=lambda url=sorted_events[event]['url']: webbrowser.open(url))
        buy_tickets_button.grid(row=event+5, column=5)
        location = sorted_events[event]['_embedded']['venues'][0]['name'] #issue with location, seems to only work if a city is used, not a state for example
        date = sorted_events[event]['dates']['start']['localDate']
        
        name_label = Label(window, text=name)
        name_label.grid(row=event+5, column=0)

        day_label = Label(window, text=date)
        day_label.grid(row=event+5, column=1)

        event_loc_label = Label(window, text=location, wraplength=400)
        event_loc_label.grid(row=event+5, column=2)
        
        min_price_label = Label(window, text=min_price)
        min_price_label.grid(row=event+5, column=3)
        
        max_price_label = Label(window, text=max_price)
        max_price_label.grid(row=event+5, column=4)

window = Tk()

search_frame = Label(window, text="Performer Name (e.g. new-york-yankees)")
search_frame.grid(row=0, column=0)
performer_entry = Entry()
performer_entry.grid(row=0, column=1, padx=5)

area_frame = Label(window, text="City (e.g. new york)")
area_frame.grid(row=1, column=0)
area_entry = Entry()
area_entry.grid(row=1, column=1, padx=5)

search_button = Button(text="Search", command=search_performers)
search_button.grid(row=0, column=2, padx=5, pady=2, rowspan=2)


event_label = Label(window, text="Event")
event_label.grid(row=3, column=0, pady=10)

date_label = Label(window, text="Date")
date_label.grid(row=3, column=1,pady=10)

loc_label = Label(window, text="Location")
loc_label.grid(row=3, column=2, pady=10)

min_label = Label(window, text="Min Price")
min_label.grid(row=3, column=3, pady=10)

max_label = Label(window, text="Max Price")
max_label.grid(row=3, column=4, pady=10)

window.mainloop()
 
```
